## FUNCTION DEFINITION 

```python
# IMPORT LIBRARY
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import f1_score, classification_report, accuracy_score,precision_score, confusion_matrix,recall_score
```

```python
# TO STORE MODEL SCORE DATA
model_scores = {}
```

```python
"""
THE FUNCTION FIND THE MODEL STANDARD DEVIATION, MEAN ACCURACY, ACCURACY, PRECISION, RECALL, F1-SCORE, ETC
"""
def measurement(model, model_info, predict_y):
        
    # FIND THE MODEL ACCURACY
    model_accuracy=accuracy_score(test_y,predict_y)
            
    # FIND THE MODEL PRECISION
    model_precision=precision_score(test_y,predict_y,average='micro')
        
    # FIND THE MODEL RECALL
    model_recall=recall_score(test_y,predict_y,average='micro')
        
    # FIND THE MODEL F1-SCORE
    model_f1_score=f1_score(test_y,predict_y,average='micro')
    
    
    # USING KFold FIND MODEL MEAN ACCURACY AND DEVIATION
    kfold = KFold(shuffle=True,n_splits=10,random_state=42)    
    model_train =cross_val_score(model, train_x, train_y, cv=kfold, scoring='accuracy')
    pd.DataFrame(model_train,columns=['Scores'])
    
    # FIND MEAN ACCURACY
    Mean_accuracy=model_train.mean()    
        
    # FIND STANDARD DEVIATION
    std=model_train.std()
    
    # DISPLAY EACH VALUES
    print('Standard Deviation:',std)
    print('Mean accuracy:',Mean_accuracy)
    print('Accuracy: ',model_accuracy)
    print('precision: ',model_precision)
    print('recall: ',model_recall)
    print('F1-Score',model_f1_score)
    
    # SAVE INTO DATAFRAME
    model_scores[f"{model_info} "] = [model_accuracy, model_precision, model_recall,model_f1_score,Mean_accuracy,std]
```

```python
test_y.shape
```

```python
"""
FUNCTION TO DISPLAY HEATMAP OF CONFUSION MATRIX OF EACH DISEASE
"""
def conf_matrix_graph(model_info,predicted):
    # PREPARE CONFESION MATRIX
    
    conf_matrix_rf = confusion_matrix(test_y, predicted)
    indices=conf_matrix_rf.shape[0]
    conf_matrix_rf = pd.DataFrame(conf_matrix_rf, 
                                  index=pd_model['Disease'].unique()[:indices], 
                                  columns=pd_model['Disease'].unique()[:indices])
    #[:42]
    # SHOW HEATMAP GRAPH
    plt.figure(figsize = (16,8))
    sns.heatmap(conf_matrix_rf, cmap="YlGnBu")
    
    # SET GRAPH LABEL
    plt.xlabel('Disease')
    plt.title(model_info)
    plt.ylabel('Disease')
    
    # RETURN THE MATRIX
    return conf_matrix_rf
```

```python
"""
Method - find list of the precaution for the patient disease
"""
def suggest_Precaution(patient_disease):
    tmp=pd_Disease_preca[pd_Disease_preca['Disease']==patient_disease]
    print('\nPrecaution:')    
    for index, row in tmp.iterrows(): 
        print(f" [1]",row['Precaution_1'])
        print(f" [2]",row['Precaution_2'])
        print(f" [2]",row['Precaution_3'])
        print(f" [2]",row['Precaution_4'])
    return tmp
```

```python
"""
Method - find list of the symptoms for the patient disease
"""

def disease_symptoms(patient_disease):
    tmp_Symptoms=[]
    
    # Find other symptoms for the disease
    tmp=pd_model[pd_model['Disease']==patient_disease][['Symptom_1','Symptom_2','Symptom_3','Symptom_4','Symptom_5','Symptom_6','Symptom_7','Symptom_8','Symptom_9','Symptom_10','Symptom_11','Symptom_12','Symptom_13','Symptom_14','Symptom_15','Symptom_16','Symptom_17']]
    for index, row in tmp.iterrows():
        for j in range(18):
            if(j>0) :
                col='Symptom_'+str(j)
                symp=str(row[col]).strip()
                if(symp!='nan'):
                    tmp_Symptoms.append(symp)
    
    # Select unique values
    tmp_Symptoms=np.unique(tmp_Symptoms)    
    return tmp_Symptoms
```

```python
"""
Method - Predict disease for the given patient symptoms
"""
def predict_disease(model,symptoms):
        
    # CREATE COPY OF PATIENT PYMPTOMS
    user_symptoms=symptoms.copy()
    
    
    # SEARCH WEIGHT FOR THE SYMPTOM
    dataSymptom = np.array(pd_symptoms["Symptom"])
    dataweight = np.array(pd_symptoms["weight"])    
    
    #for j in range(17):
    for j in range(6):
        if (j <len(symptoms)):
            for k in range(len(dataSymptom)):
                if symptoms[j]==dataSymptom[k]:
                    symptoms[j]=dataweight[k]
        else:
            symptoms.append(0)
    # SHOW SYMPTOMS WEIGHT    
    print ('Patient symptoms:',user_symptoms)
    print('\nPatient symptoms weight:',symptoms)    
    
    # PREDICT DISEASE
    symptoms = [symptoms]      
    predicted_disease = model.predict(symptoms)

    # SHOW PREDICTED DISEASE
    print('\nDisease : ',predicted_disease)        
    
    
    # SHOW ANY OTHER SYMPTOMS FOR THE DISEASE
    disease_Symptoms=disease_symptoms(predicted_disease[0])    
    print ('\nDisease symptoms (all):',disease_Symptoms,'\n')
    
    # REMOVE USER SYMTOMS FOR DISEASE SYMPTOMS
    print ('\nMore symptoms:')
    for s in disease_Symptoms:            
        if(s in user_symptoms) or s=='nan':      
            tmp=0
        else:
            print(' > ', s) 
            
    # DISEASE PRECAUTIONS
    suggest_Precaution(predicted_disease[0])
```
